```python
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
```

# Age and Survival
>We plot the age and death to see if there is any obvious correlation.

```python
df_train = pd.read_csv('data/train.csv')

survived = df_train['Survived'].values
age = df_train['Age'].values

reds_dead = []
blues_survive = []

for i in range(len(survived)):
    if survived[i] == 1:
        if np.isnan(age[i]) == False:
            blues_survive.append(age[i])
    else:
        if np.isnan(age[i]) == False:
            reds_dead.append(age[i])

red_ = reds_dead[0:289]
red__ = reds_dead[289:]

plt.margins(0.05)
plt.xlabel('Index')
plt.ylabel('Age')
plt.plot(red_, 'ro')
plt.plot(red__, 'ro')
plt.plot(blues_survive, 'bo')
plt.show()
```

>There doesn't seem to be any strong correleation with age.

> **Note:** We will revisit this below.

# Gender and Survival
>We plot gender and survival to see if there is any strong correlation. 

```python
sex = df_train['Sex'].values

red_female = []
red_male = []

blue_female = []
blue_male = []

for i in range(len(survived)):
    if survived[i] == 1:
        if sex[i] == 'female':
            blue_female.append(1)
        else:
            blue_male.append(1)
    else:
        if sex[i] == 'female':
            red_female.append(0)
        else:
            red_male.append(0)

n_groups = 2

rm = len(red_male)
bm = len(blue_male)
rf = len(red_female)
bf = len(blue_female)

means_red = (rm, rf)
means_blue = (bm, bf)
 
# create plot
fig, ax = plt.subplots()
ax.margins(0.04, 0.00)
index = np.arange(n_groups)
bar_width = 0.35
opacity = .8

error_config = {'ecolor': '0.3'}
 
rects1 = plt.bar(index, means_red, bar_width,
                 alpha=opacity,
                 color='r',
                 label='Died')
 
rects2 = plt.bar(index + bar_width, means_blue, bar_width,
                 alpha=opacity,
                 color='b',
                 label='Survived')
 
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Death Count by Gender')
plt.xticks(index + bar_width, ('Male', 'Female'))
plt.legend()
 
plt.tight_layout()
plt.show()

print("Values", '\n', "Male:", rm, bm, '\n', 'Female:', rf, bf, '\n')
print("Probability of survival", '\n', 'Male:', '%.2f' % float(bm/(rm + bm)), '\n', 'Female:', '%.2f' % float(bf/(rf+bf)))
```

>There seems to be some promise when comparing these two.

# Fare and Survival
>We plot fare price and survival to see if there is any strong correlation.

```python
fare = df_train['Fare'].values

reds_fare = []
blues_fare = []

for i in range(len(survived)):
    if survived[i] == 1:
        blues_fare.append(fare[i])
    else:
        reds_fare.append(fare[i])

red_ = reds_fare[0:289]
red__ = reds_fare[289:]

plt.margins(0.05)
plt.xlabel('Index')
plt.ylabel('Fare Price')
plt.plot(red_, 'ro')
plt.plot(red__, 'ro')
plt.plot(blues_fare, 'bo')
plt.show()
```

>Not a strong correlation, but seems to be some relationship with a higher fare price and survival rate.

> **BUT** lets do a bar graph to get a better idea.

```python
high_fare_survive = []
high_fare_dead = []

for i in range(len(fare)):
    if fare[i] > 70.0:
        if survived[i] == 1:
            high_fare_survive.append(survived[i])
        else:
            high_fare_dead.append(survived[i])
            
    
    
n_groups = 1

rfare = len(high_fare_dead)
bfare = len(high_fare_survive)

means_red = (rfare)
means_blue = (bfare)
 
# create plot
fig, ax = plt.subplots()
ax.margins(0.04, 0.00)
index = np.arange(n_groups)
bar_width = 0.35
opacity = .8

error_config = {'ecolor': '0.3'}
 
rects1 = plt.bar(index, means_red, bar_width,
                 alpha=opacity,
                 color='r',
                 label='Died')
 
rects2 = plt.bar(index + bar_width, means_blue, bar_width,
                 alpha=opacity,
                 color='b',
                 label='Survived')
 
plt.xlabel('Red: dead Blue: survive')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
    
print(len(high_fare_survive))
print(len(high_fare_dead))
```

>There is a larger probability of survival given the fare price.

# Random Forests
>As you might guess, it is a bit time consuming to shift through different features in the dataset. What if we have a dataset with 10000 different features. Plotting these would take quite sometime.

>A simple way to find good candidates for features would be to use the ensemble method, **Random Forest Classifier**.

>The script below implements a Random Forest Classifier.

```python
import sklearn

import pandas as pd
import numpy as np

from sklearn.ensemble import RandomForestClassifier

#NOTE: SCORE=.73206 on titanic scoreboard

#Process the data
def processData(df, train=True):
    df = df.drop(['Name', 'Ticket', 'Cabin', 'Embarked'], axis=1)

    mean_age = df['Age'].mean()
    df['Age'] = df['Age'].fillna(mean_age)

    fare_means = df.pivot_table('Fare', index='Pclass', aggfunc='mean')
    df_test['Fare'] = df_test[['Fare', 'Pclass']].apply(lambda x:
                            fare_means[x['Pclass']] if pd.isnull(x['Fare'])
                            else x['Fare'], axis=1)

    df['Gender'] = df['Sex'].map({'male' : 0, 'female' : 1}).astype(int)

    df = df.drop(['Sex'], axis=1)

    if train == True:
        cols = df.columns.tolist()
        cols = [cols[1]] + cols[0:1] + cols[2:]
        df = df[cols]

    return df.values


df_train = pd.read_csv('data/train.csv')
df_test = pd.read_csv('data/test.csv')

train_data = processData(df_train)
print('\n')
test_data = processData(df_test, False)
    
clf = RandomForestClassifier(n_estimators=200)
clf.fit(train_data[0:, 2:], train_data[0:, 0])

output = clf.predict(test_data[:,1:])

result = np.c_[test_data[:,0].astype(int), output.astype(int)]
df_result = pd.DataFrame(result[:,0:2], columns=['PassengerId', 'Survived'])

print(df_result[:10])
```

>We dropped the 'Name', 'Ticket', 'Cabin', and 'Embarked' categories for simplicity. Then we converted the missing 'Age' nan values to the mean of ages and changed the 'Sex' to an integer... male : 0 or female : 1.  

```python
feature_titles = np.array(['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Gender'])

important_features = clf.feature_importances_
print(important_features)
```

```python
sort_least_to_great = np.argsort(important_features)
print(sort_least_to_great)
print(important_features[sort_least_to_great])
print(feature_titles[sort_least_to_great])
```

>So these are our ranked features.

>We can also plot them to see relative importance.

```python
padding = np.arange(len(important_features)) + .5
plt.barh(padding, important_features[sort_least_to_great], align='center', color='orange')
plt.yticks(padding, feature_titles[sort_least_to_great])
plt.xlabel("Relative Importance")
plt.show()
```

> We have three decent features from above. These are the three features that we plotted above. So, just by intuition we picked three good features to test.

>However, we initially thought 'Age' was not a great feature from just a plot itself. Can plots be trusted in all honesty then? The answer really depends. Good visualization are a must, but too many data points and the choice of the wrong plot (scatter point plot!) can be deceiving.

> **NOTE:** It turns out that keeping Pclass in our features gives a boost to the accuracy. So, this gives somewhat of a guide relative importane accuracy. 

# Takeaways

> 1. **Bar graphs for features**, or some type of plot that will distinctly show any outliers/good features.


> 2. **Random Forests are a nice way to find some good features out of the gate**. As a matter of fact, Random Forests are a good out-of-the-bag classifier to try first. They are a bit of a level up from Naive Bayes, KNN, Logisitic Reg., ect., but still something worth considering right out of the gate.


> 3. **Relative Importance is well, relative**. Pclass gave, on average, a boost to the accuracy (on the Kaggle submission).
